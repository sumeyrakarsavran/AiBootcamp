```python
import numpy as np #lineer cebir vs
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.arange(20) #0dan 19a kadar ardisik rakamlar
y=np.random.normal(10,1,20) #ort 10, st sapma 1 olan 20 rastgele rakam
z=np.random.normal(10,2,20) #ort 10, st sapma 2 olan 20 rastgele rakam #normal destribution için normal yazdık
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('x degerleri')
plt.ylabel('y ve z degerleri')
plt.title('x in y ve z ye göre değişimi')
```

```python
fig=plt.figure(figsize=(10,6)) #10inch genisliğinde 6 inch yuksekliğinde olucak
ax=fig.add_subplot(131) #1 satır 3 sütunlu grafiğin 1.sutünu
ax.plot(x,y,c='red',marker='d') #çizgi grafik, c color demek, marker diamond olsun
ax.set_title('x vs y')

bx=fig.add_subplot(132) #1 satır 3 sütunlu grafiğin 2.sutünu
bx.scatter(x,z,marker='*') #nokta grafik

cx=fig.add_subplot(133) 
cx.hist(y); #histogram grafik ; koyunca sayılı çıktı gidiyor
```

### feature engineering

```python
df['Date']=pd.to_datetime(df['Date']) #date in type ı objectti artık date oldu
```

```python
df.info()
```

```python
#day, month, year diye df'de yeni sütunlar oluşturuyoruz
df['day']= (df['Date']).dt.day
df['month']= (df['Date']).dt.month
df['year']= (df['Date']).dt.year 
```

```python
df.head()
```

```python
df['month_name']=df['Date'].dt.month_name() #sayıya göre ayı bulup ismini getiriyo işlem yaptığı için fonksiyondur
```

```python
df['weekday']=df['Date'].dt.day_name()
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
df['Hour']=(df['Time']).dt.hour #df nin başında parantez olmasa da sorun değil galiba
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperature = [10, 12, 15, 18, 21, 24, 27, 26, 23, 19, 14, 11]

# Convert data to a dataframe
df = pd.DataFrame({'Temperature': temperature}, index=months)

# Create the polar chart using the dataframe
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(np.radians(np.arange(0, 360, 30)), df['Temperature'])
ax.fill(np.radians(np.arange(0, 360, 30)), df['Temperature'], alpha=0.5)
ax.set_xticks(np.radians(np.arange(0, 360, 30)))
ax.set_xticklabels(months);
```

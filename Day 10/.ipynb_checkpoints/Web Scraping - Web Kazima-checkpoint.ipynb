```python
#tüm yazılarının başlıklarının linki bu article olanlarda
```

```python
page=requests.get('https://www.karar.com/yazarlar/zafer-acar') #tüm yazıların old sayfa
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
linkler=soup.find_all('article', {'class':'item box-shadow'}) #bir değil bütün linkleri almak için
```

```python
linkler
```

```python
yazilar=[]
basliklar=[]
for link in linkler:
    print('https://www.karar.com'+ link.a.get('href')) #tüm başlıkların linki
```

```python
yazilar=[]
basliklar=[]
for link in linkler:
    link1='https://www.karar.com'+link.a.get('href')
    print(link.h3.text) #tüm başlıklar
```

```python
yazilar=[]
basliklar=[]
for link in linkler:
    link1='https://www.karar.com'+link.a.get('href')
    baslik=link.h3.text
    page2=requests.get(link1) #her başlığın ayrı ayrı sayfası
    soup2=BeautifulSoup(page2.content,'html.parser')
    makale=soup2.select('article.article-content')[0].text
    yazilar.append(makale)
    basliklar.append(baslik)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=['Title','Article' ])
```

```python
df['Title' ]=basliklar
df['Article']=yazilar
```

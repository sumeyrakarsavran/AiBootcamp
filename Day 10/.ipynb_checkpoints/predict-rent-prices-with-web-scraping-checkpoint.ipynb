```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.zingat.com/samsun-merkez-atakum-korfezde-kiralik-2-0-daire-5147042i"
```

```python
page = requests.get(url)
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser') #html i okunablir yapıyo
```

```python
soup.prettify
```

```python
soup.find_all('h3')
```

```python
soup.select('feature-item feature-price').text #türü.class adı [0].text yazılır
```

```python
price_element = soup.select('div.feature-price')

# Tek bir öğeyi almak ve metni çekmek için [0] ile erişin
if price_element:
    print(price_element[0].text.strip())  # Örneğin, "9.500 TL"
else:
    print("Öğe bulunamadı.")
```

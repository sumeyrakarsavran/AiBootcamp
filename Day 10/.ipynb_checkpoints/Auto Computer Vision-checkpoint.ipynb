```python
import cv2
import matplotlib.pyplot as plt
import cvlib
from cvlib.object_detection import draw_bbox

#80.000 nesneyi tanÄ±yabiliyor bu paket
```

```python
img=cv2.imread('cars.jpg')
```

```python
plt.imshow(img)
```

```python
box, label, count=cvlib.detect_common_objects(img) #yolov4, yolov3 kuruyo
```

```python
output=draw_bbox(img,box, label, count)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(output)
```

```python
label.count('car')
```

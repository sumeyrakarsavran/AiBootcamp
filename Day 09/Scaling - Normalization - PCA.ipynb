# Scaling - Normalization - PCA

```python
import pandas as pd
```

```python
df=pd.read_csv('pima-indians-diabetes.csv')
#hamilelik döneminde geçici şeker hastalığı oluyo, 1se şeker geçici hastlaığı var 0sa yok
```

```python
df.head()
```

z-skor yapmak, normalize etmek, scale etmek
tüm rakamları 0-1 arasına çekmek oluyor

şu ana kadarki projeleri 0-1 arasına çekip yapsaydık daha başarılı olurduk

```python
from sklearn.preprocessing import normalize, scale
```

```python
x=df.drop('Outcome', axis=1) #outcome tahmin edileceği için onu çıkarıcaz
```

```python
x.head()
```

```python
yeni_x=scale(x) # -1le 1 arasına alır
```

```python
yeni_x
```

```python
pd.DataFrame(yeni_x)
```

```python
normalizeddata=normalize(x) #0-1 arasına aldı, scale de edilebilir normalize da edilir
```

```python
pd.DataFrame(normalizeddata)
```

### artık scale veya normalize ettikten sonra modeli eğit

# Boyut Kucultme PCA - Principle Component Analysis

```python
# 5bin sütun vs varsa boyut küçültebilirsin
```

```python
#bu veride 8 sütun vardı 2 sütuna şöyle indirebiliriz
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(3)
```

```python
x.head()
```

```python
data2=pca.fit_transform(x)
```

```python
data2
```

```python
pd.DataFrame(data2) #veriyi 2 sütuna indirdik. aynı veri bu şu an
```

```python
pd.DataFrame(data2, columns=['PCA1','PCA2','PCA3'])
```

```python
pca. explained_variance_ratio_ #bunun ne kadarını temsil etmiş data ilk ve ikinci sütunda
```

```python
sum(pca.explained_variance_ratio_) # kayıp olmuş mu diye bakınca yüzde 95i duruyo 5i kayıp
```

```python
import pandas as pd
```

```python
df=pd.read_csv('pima-indians-diabetes.csv') 
#indian kızılderili demek kızılderililerin pima kabilesindeki kadınların hamilelik döneminde geçici şeker hastası olma durumu
```

```python
df.head()
```

boyutu matrixden de fazlaysa buna tensor diyoruz

```python
df.isnull().sum()
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow. keras. layers import Dense
```

```python
x=df.drop('Outcome', axis=1)
y=df [ ['Outcome' ] ]
```

```python
df.columns
```

```python
#aynı şey
x=df [['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
y=df[['Outcome' ]]
```

```python
#üçüncü farklı yöntem
#x=df.iloc[:,0:8]
#y=df.iloc[:,8]
```

```python
x.head()
```

```python
y.head()
```

```python
#katman ekliyoruz toplamda 6 katman ekledik, sayıları kafamıza göre yapıyoruz
#nöronların fazladan başlayıp azalması lazım
model=Sequential()
model.add(Dense(120, activation='relu')) #layer eklemek oluyor
model.add(Dense(80, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1,activation='sigmoid')) #binary olması için sonuç
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
#loss yani hata, sonuç 0 veya 1 olcak
#adam, aşağıdaki fotoda sağdaki grafikteki en küçük değeri bulmaya yaklaştıkça iterasyonu azaltmaya denir
```

```python
df.shape #en az bin olmalıydı
```

```python
model.fit(x,y,epochs=250, batch_size =32, validation_split=.20, verbose=1)

#epochs, bir sokaktan kaç kez geçeceği
#epochs, insanın aynı sokaktan sürekli geçince o sokak hakkında bilgisinin artması gibi bir şeydir.
#epochs, nöronların arasında git gel yapa yapa yeni öğrendiği bilgilerle weightleri update ediyor

#batch_size eşit parçalara bölüp çalışır. 768i azalta azalta yapıyor bilgisayar zorlanır
#32 yazarsak 768/32 yani her seferinde 24 satır alacak

#veriyi ikiye bölüyoruz, 0.8 eğitim, 0.2 test için

#verbose 1 se her epoch sonucu yazar, 0 ise model kendini eğitene kadar bırakmıyor
```

```python
model.summary()
```

```python
from sklearn.preprocessing import normalize, scale
```

```python
x=scale(x)
```

```python
history = model.fit(x,y,epochs=250, batch_size =32, validation_split=.20, verbose=1)
#train test diye ayırmadığımız için accuracy 1 çıkıyo 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy' ], label='Accuracy')
plt.plot(history.history['val_accuracy' ], label='Val_Accuracy' )
plt.legend()
```

```python
# DNN - Regression
```

```python
import pandas as pd
```

```python
df=pd.read_pickle('kc_house.pkl') #pickle yapılmış olarak saklanmış bir data
#yani feature engineering yapılmış
```

```python
df.head()
```

```python
x=df.drop('price', axis=1)
y=df [ ['price' ] ]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y, test_size =.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(120,activation='relu'))
model.add(Dense(80,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1)) #tek bir data alacağımız için sigmoid yazmayacağız
model.compile(loss='mse', optimizer='adam') #mse kullancağımız için accuracy kullanmayacağız
```

```python
x.shape
```

```python
model.fit(x_train,y_train, validation_data=(x_test,y_test),batch_size=128,epochs=300)
```

```python
tahmin=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_test,tahmin) #0.8 üstü başarılı demek, epoche 1500 yapınca başarı oranı 0.9 üstü olur
```

```python
mean_squared_error(y_test, tahmin) **.5 #100bin dolar altı başarı demek(?) , epoche 1500 yapınca mse 50bin altı olur
```
